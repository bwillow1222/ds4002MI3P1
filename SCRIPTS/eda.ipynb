```python
# Download necessary packages
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/bwillow1222/ds4002MI3P1/main/DATA/Scored_Data.csv'
data = pd.read_csv(url)
```

```python
import matplotlib.pyplot as plt

# Group the data by Speaker and AtDebate, and count the number of entries for each
grouped_data = data.groupby(['Speaker', 'AtDebate']).size().unstack(fill_value=0)

# Create the bar graph
grouped_data.plot(kind='bar', figsize=(10, 6))

# Add labels and title
plt.title('Count of Entries by Speaker Grouped by Debate Participation')
plt.xlabel('Speaker')
plt.ylabel('Count of Entries')
plt.legend(title="At Debate", labels=["Not at Debate", "At Debate"])
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
