This notebook contains two scripts that generate bar graphs:

1. Bar Graph of Entries by Speaker and Debate Participation:
   - Groups the data by speaker and whether they participated in the debate.
   - Displays a bar graph showing the count of entries for each speaker.

2. Bar Graph of Entries for Each Topic:
   - Counts the number of entries for each topic (e.g., Economy, Immigration, etc.).
   - Displays a bar graph of these counts.

```python
# Download necessary packages
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

url = 'https://github.com/bwillow1222/ds4002MI3P1/raw/main/DATA/processed/Scored_Data.csv'
data = pd.read_csv(url)

data.head()
```

```python

# Group the data by Speaker and AtDebate, and count the number of entries for each
grouped_data = data.groupby(['Speaker', 'AtDebate']).size().unstack(fill_value=0)

# Create the bar graph
grouped_data.plot(kind='bar', figsize=(10, 6))

# Add labels and title
plt.title('Count of Entries by Speaker Grouped by Debate Participation')
plt.xlabel('Speaker')
plt.ylabel('Count of Entries')
plt.legend(title="At Debate", labels=["Not at Debate", "At Debate"])
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python

# Group the data by 'Topic' and count the number of entries in each category
topic_counts = data['Topic'].value_counts()

# Plot the count of each topic as a bar graph
plt.figure(figsize=(10, 6))
topic_counts.plot(kind='bar')

# Add labels and title
plt.title('Count of Entries for Each Topic Category')
plt.xlabel('Topic')
plt.ylabel('Count of Entries')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Replace NaN or missing values in the 'Topic' column with 'None'
data['Topic'] = data['Topic'].fillna('None')

# Group the data by Topic and Speaker (Candidate), and count the number of entries for each
grouped_data = data.groupby(['Topic', 'Speaker']).size().unstack(fill_value=0)

# Create a stacked bar graph for Topics, with bars stacked by Candidate (Speaker)
grouped_data.plot(kind='bar', stacked=True, figsize=(10, 6))

# Add labels and title
plt.title('Number of Comments by Topic with Stacked Candidates (Including None)')
plt.xlabel('Topic')
plt.ylabel('Number of Comments')
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()

```
